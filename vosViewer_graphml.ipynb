```python
import networkx as nx

journal=nx.read_graphml("data/journal.graphml")
```

```python
journal.nodes.data()
```

```python
descriptionText=[x[1]['description'].split('d>')[3].split('</')[0] for x in journal.nodes(data=True)]
issnText=[x[1]['description'].split('td>')[7].split('<')[0] for x in journal.nodes(data=True)]

descriptionMap={node:value for node,value in zip(list(journal),descriptionText)}
issnMap={node:value for node,value in zip(list(journal),issnText)}
citationsMap=nx.get_node_attributes(journal, 'weight<citations>')
nx.set_node_attributes(journal, descriptionMap, name="description")
nx.set_node_attributes(journal, issnMap, name="issn")
nx.set_node_attributes(journal, citationsMap, name="citations")
```

```python
journal.nodes.data()

nx.write_graphml(journal,"data/journal.graphml")
nx.write_gml(journal,"data/journal.gml")
nx.write_gexf(journal,"data/journal.gexf")
```

```python
attrs=['weight<links>','weight<total_link_strength>','weight<citations>','score<journal_impact_factor>','size','r','g','b' ]
for attr in attrs:
    for node in journal:
        try:
            del journal.nodes[node][attr]
        except:
            pass
```
